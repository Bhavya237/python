```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r'C:\Users\Bhavya\Downloads\food delivery costs.csv')
```

```python
data.info()
```

```python
data['Order Date and Time']=pd.to_datetime(data['Order Date and Time'])
```

```python
data['Delivery Date and Time']=pd.to_datetime(data['Delivery Date and Time'])
```

```python
data.info()
```

```python
def extract(value):
    a=str(value).split(' ')
    return a[0]

data['Discounts and Offers']=data['Discounts and Offers'].apply(extract)
data.head()
```

```python
def remove(value):
    if '%' in value:
        a=value.replace('%',' ')
        return float(a)
    else:
        return float(value)

data['Discounts and Offers']=data['Discounts and Offers'].apply(remove)
data.head()
```

```python
data['Discounts and Offers']=data['Discounts and Offers'].fillna(0)
data.head()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
data.loc[(data['Discounts and Offers']<=15),'Discounts and Offers']=(data['Discounts and Offers']/100)*data['Order Value']
data.head()
```

```python
data['cost']=data['Delivery Fee']+data['Payment Processing Fee']+data['Discounts and Offers']
data.head()
```

```python
data['Profit']=data['Commission Fee']-data['cost']
data.head()
```

```python
Cost_distibution=data[['Discounts and Offers','Delivery Fee','Payment Processing Fee']].sum()
Cost_distibution
```

```python
plt.figure(figsize=(3,3))
plt.pie(Cost_distibution,labels=Cost_distibution.index,autopct='%1.2f%%')
plt.show()
```
